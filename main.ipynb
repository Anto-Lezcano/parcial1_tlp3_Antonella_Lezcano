Actividad 3- Limpieza y normalizacion de datos                     
Se utilizaran metodos de pandas para realizar la limpieza de datos para su analisis.
- 3.1. Eliminación de filas, cubrir valores con media:                  
Para valores numericos se remplazaran los valores con la media                          
Para datos de tipo string se eliminaran las filas con datos del tipo Nan


```python
df.dropna(inplace=True)
```

- 3.2: La columna Fecha deberan pasarla al tipo datetime con su metodo correspondiente.

```python
serie = pd.Series(df['Fecha'])

serie_1 = pd.to_datetime(serie, format='%Y-%m-%d', errors='coerce')
serie_2 = pd.to_datetime(serie, format='%d-%m-%Y', errors='coerce')

serie_final = serie_1.combine_first(serie_2)

df['Fecha']= serie_final
```

- 3.3: Corregir las mayusculas en el caso de Nombre y Apellido (Si es que corresponde.)

```python
df['Apellido_Cliente']= df['Apellido_Cliente'].str.title()
df['Nombre_Cliente']= df['Nombre_Cliente'].str.title()
```

Actividad 4- Estadisticas descriptivas                         
Se utilizaran metodos de pandas para obtener las estadisticas descriptiva
- 4.1: calcular nuevamente las estadísticas descriptivas para observar el impacto del proceso de limpieza en los datos numéricos.

```python
df.describe()
```

- 4.2: Calcular estadísticas específicas por grupo (Agrupar dos columnas).

```python
df.groupby(['Precio_Unitario', 'Total_Venta']).mean(numeric_only=True)
```

Actividad 5- Visualizacion de datos                           
Se hara uso de matplotlib para realizar el grafico de barras

```python
ventas_product = df.groupby('Region')['Total_Venta'].sum()

plt.Figure(figsize=(4,5))
plt.bar(ventas_product.index, ventas_product.values, edgecolor='black', color='gray')
plt.title('Analisis de ventas por Region', fontsize=15)
plt.tight_layout()
plt.xlabel('Region')
plt.ylabel('Ventas')
plt.xticks(rotation=45)

plt.show()
```

Actividad 6- Exportacion y consulta con squelite           
Se hara uso de pandas y sqelite3 para realizar las consultas y la exportacion.

```python
import sqlite3 as sq

conection = sq.connect('ventas.db')

df.to_sql('ventas', conection, if_exists='replace', index=False)

consulta = pd.read_sql_query('SELECT Producto, Precio_Unitario FROM ventas LIMIT 5', conection)

print(consulta)

conection.close()
```
